Exercício 1 - App de vendas para uma empresa X que vende em atacado e oferece descontos maiores dependendo da quantidade de produto comprado.

```python
print('Bem vindo a Loja da Marcela Almeida Ferrari')
preco = float(input('Qual o valor do produto? '))
qtd = int(input('Informe a quantidade: '))
valor_total = preco * qtd # variável para calcular o valor total, de acordo com a quantidade
if qtd <= 9: # sem desconto
    valor_final = valor_total
elif qtd >= 10 and qtd <= 99: # desconto de 5%
    valor_final = valor_total * 0.95 # cálculo para apresentar o valor com desconto
elif qtd >= 100 and qtd <= 999: # desconto de 10%
    valor_final = valor_total * 0.90
else: # desconto de 15%
    valor_final = valor_total * 0.85
print('Valor sem desconto: {:.2f}. Valor com desconto: {:.2f}.'.format(valor_total, valor_final))
```

Exercício 3 - App para uma empresa de logística que calcula valores de acordo com a rota escolhida, dimensões e peso do objeto.

```python
#-----------COMEÇO DA FUNÇÃO DIMENSOESOBJETO-----------
def dimensoesObjeto():
  global volume_m # instrução global para o multiplicador do volume
  while True:
    try:
      altura = float(input('Qual a altura do objeto? (em cm) '))
      comprimento = float(input('Qual o comprimento do objeto? (em cm) '))
      largura = float(input('Qual a largura do objeto? (em cm) '))
      volume = altura * comprimento * largura
      print('O volume do objeto é {}cm³'.format(volume))
      if (volume < 1000): # em cada if e elif será calculado o valor de acordo com o volume
        volume_m = 10
        return volume_m
      elif (1000 <= volume < 10000):
        volume_m = 20
        return volume_m
      elif (10000 <= volume < 30000):
        volume_m = 30
        return volume_m
      elif (30000 <= volume < 100000):
        volume_m = 50
        return volume_m
      else:
        print('Objetos com volume a partir de 100000cm³ não são aceitos')
        continue
    except ValueError: # caso o usuário digite um caractere não numérico, a exceção será tratada e voltará para o início do loop
      print('Você digitou um valor não numérico.')
      continue
#-----------FIM DA FUNÇÃO DIMENSÕESOBJETO--------------
#-------------COMEÇO DA FUNÇÃO PESOOBJETO--------------
def pesoObjeto():
  global peso_m # instrução global para o multiplicador do peso
  while True:
    try:
      peso = float(input('Qual o peso do objeto? (em kg) '))
      if (peso <= 0.1): # calculando o valor de acordo com o peso
        peso_m = 1
        return peso_m
      elif (0.1 < peso <= 1):
        peso_m = 1.5
        return peso_m
      elif (1 < peso <= 10):
        peso_m = 2
        return peso_m
      elif (10 < peso <= 30):
        peso_m = 3
        return peso_m
      else:
        print('Objetos com peso acima de 30kg não são aceitos.')
        continue
    except ValueError:
      print('Você digitou um valor não numérico.')
      continue
#---------------FIM DA FUNÇÃO PESOOBJETO---------------
#-------------COMEÇO DA FUNÇÃO ROTAOBJETO--------------
def rotaObjeto():
  global rota_m # instrução global para o multiplicador da rota
  print('Informe a rota desejada:')
  print('RS - De Rio de Janeiro até São Paulo')
  print('SR - De São Paulo até Rio de Janeiro')
  print('BS - De Brasília até São Paulo')
  print('SB - De São Paulo até Brasília')
  print('BR - De Brasília até Rio de Janeiro')
  print('RB - Rio de Janeiro até Brasília')
  while True:
    rota = input()
    if (rota == 'RS') or (rota == 'SR'): # calculando o valor de acordo com a rota
      rota_m = 1
      return rota_m
    elif (rota == 'BS') or (rota == 'SB'):
      rota_m = 1.2
      return rota_m
    elif (rota == 'BR') or (rota == 'RB'):
      rota_m = 1.5
      return rota_m
    else: # se for digitado algo diferente das siglas, inicia novamente o loop
      print('Sigla não encontrada. Tente novamente')
      continue
#---------------FIM DA FUNÇÃO ROTAOBJETO---------------

#------------------PROGRAMA PRINCIPAL------------------
print('Bem vindo a Companhia de Logística Marcela Almeida Ferrari')
print('Total a pagar: R${:.2f} (dimensões {} * peso {} * rota {})'.format(dimensoesObjeto() * pesoObjeto() * rotaObjeto(), volume_m, peso_m, rota_m))
```

Exercício 4 - Software de controle de estoque para uma bicicletaria. Cadastra e remove peças, consulta por código ou fabricante.

```python
listaPecas = []
#------------COMEÇO DA FUNÇÃO CADASTRARPECA------------
def cadastrarPeca(codigo):
  print('Você escolheu a opção Cadastrar Peça')
  print('Código da peça: {:03d}'.format(codigo))
  nome = input('Digite o nome da peça: ')
  fabricante = input('Digite o fabricante da peça: ')
  valor = float(input('Digite o valor da peça: '))
  dicionarioPecas = {'codigo' : codigo, # criando o dicionário
                     'nome'   : nome,
                     'fabricante' : fabricante,
                     'valor'  : valor}
  listaPecas.append(dicionarioPecas.copy()) # adicionando uma cópia do dicionário na lista
#-------------FIM DA FUNÇÃO CADASTRARPECA--------------
#------------COMEÇO DA FUNÇÃO CONSULTARPECA------------
def consultarPeca():
  while True:
    try:
      print('Você escolheu a opção Consultar Peças')
      consulta = int(input('Informe a opção desejada:\n' # menu para escolher o modo de consulta
                          '1 - Consultar todas as peças\n'
                          '2 - Consultar peças por código\n'
                          '3 - Consultar peças por fabricante\n'
                          '4 - Retornar\n'))
      if (consulta == 1): # consultar todas as peças
        for peca in listaPecas:
          for key, value in peca.items():
            print('{} : {}'.format(key, value))
      elif (consulta == 2): # consulta por código
        consulta_cod = int(input('Digite o código da peça: '))
        for peca in listaPecas:
          if (peca['codigo'] == consulta_cod):
            for key, value in peca.items():
              print('{} : {}'.format(key, value))
      elif (consulta == 3): # consulta por fabricante
        consulta_cod = input('Digite o fabricante da peça: ')
        for peca in listaPecas:
          if (peca['fabricante'] == consulta_cod):
            for key, value in peca.items():
              print('{} : {}'.format(key, value))
      elif (consulta == 4): # retorna ao menu anterior
        break
      else: # caso digite números diferentes do menu, volta ao início do loop
        print('Número não encontrado no menu. Tente novamente')
        continue
    except ValueError: # se for digitado um valor não inteiro ou não numérico (nos campos que exigem valores inteiros)
      print('Você digitou um valor não inteiro ou não numérico.')
#-------------FIM DA FUNÇÃO CONSULTARPECA--------------
#-------------COMEÇO DA FUNÇÃO REMOVERPECA-------------
def removerPeca():
  print('Você escolheu a opção Remover Peça')
  remover = int(input('Digite o código da peça que deseja remover: '))
  for peca in listaPecas:
    if (peca['codigo'] == remover):
      listaPecas.remove(peca)
#--------------FIM DA FUNÇÃO REMOVERPECA---------------

#------------------PROGRAMA PRINCIPAL------------------
print('Bem vindo ao Controle de Estoque da Bicicletaria da Marcela Almeida Ferrari')
registro = 0
while True:
  try:
    opcao = int(input('Informe a opção desejada:\n1 - Cadastrar peça\n' # menu de opções
                      '2 - Consultar peças\n3 - Remover peça\n4 - Sair\n'))
    if (opcao == 1): # cadastrar peça
      registro += 1
      cadastrarPeca(registro)
    elif (opcao == 2): # consultar peças
      consultarPeca()
    elif (opcao == 3): # remover peças
      removerPeca()
    elif (opcao == 4): # encerra o programa
      print('Saindo do programa...')
      break
    else:
      print('Número não encontrado no menu. Tente novamente')
      continue
  except ValueError:
    print('Você digitou um valor não inteiro ou não numérico.')
```
